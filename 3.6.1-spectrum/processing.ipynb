```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

```python
def line(x, k, b):
    return x*k + b
```

```python
df = pd.read_csv(r"raw-data/data.csv")

rev_nu_0 = df['rev_nu_0']
tau = df['tau']

delta_rev_nu_0 = df['Delta_rev_nu_0']
```

```python
opt, cov = curve_fit(line, tau, rev_nu_0)
```

```python
plt.figure(figsize=(8, 4.5), dpi=200)

plt.minorticks_on()
plt.grid(which='minor', linestyle=':', color='0.9')
plt.grid(which='major', linestyle='-', color='0.8', linewidth=0.3)

plt.errorbar(tau, rev_nu_0, yerr=delta_rev_nu_0, linestyle='', marker='.', markersize=10, label="Experiment", color="tab:blue", alpha=0.3)
plt.plot(tau, line(tau, opt[0], opt[1]), label=r"Fitting", color="tab:blue", linestyle='-')

plt.ylabel(r"$1/\nu_0$, мкс", fontsize=14)
plt.xlabel(r"$\tau$, мкс", fontsize=14)

plt.legend()
plt.savefig(r"images\data.png", facecolor="white")
plt.show()
```
