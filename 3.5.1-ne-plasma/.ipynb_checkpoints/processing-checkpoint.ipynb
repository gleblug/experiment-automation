```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.txt")
U = df['U']
I = df['I']
```

```python
plt.figure(figsize=(10, 6), dpi=200)

plt.errorbar(I*6/155, U, xerr=6/155, yerr=0.003*U, linestyle='', marker='.')
plt.xlabel("Сила тока, мА")
plt.ylabel("Напряжение, В")

plt.legend()
plt.show()
```
