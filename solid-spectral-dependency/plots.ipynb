```python
# %load /home/glebi/git/experiment-automation/processing_tools.py
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
import scienceplots

plt.style.use(['science', 'russian-font'])
styles = plt.style.available

matplotlib.rcParams.update({
    'figure.figsize': [6, 4],
    'savefig.facecolor': 'white',
    'figure.dpi': 150.0,
    'font.size': 12.0,
    'savefig.dpi': 250
})
```

```python
df = pd.read_csv("data.csv", sep="\t")

U = df["U"]
hv = df["hv, мэВ"]
```

```python
plt.plot(hv, U, ".--", markersize=12, lw=1)
plt.xlabel(r"Энергия кванта $h\nu$, мэВ")
plt.ylabel(r"Сигнал фотоответа $\Delta \Sigma/N_0$")
plt.savefig("A.png")
plt.show()
```

```python
plt.plot(hv, U**.5, ".--", markersize=12, lw=1)
plt.xlabel(r"Энергия кванта $h\nu$, мэВ")
plt.ylabel(r"Сигнал фотоответа $(\Delta U/N_0)^{1/2}$")
plt.savefig("B.png")
plt.show()
```

```python
plt.style.use(['no-latex'])

for style in styles:
    matplotlib.rcParams.update({
        'figure.figsize': [6, 4],
        'savefig.facecolor': 'white',
        'figure.dpi': 150.0,
        'font.size': 12.0,
        'savefig.dpi': 250
    })
    plt.style.use([style])
    
    plt.clf()
    plt.subplots_adjust(bottom=0.15)
    plt.plot(hv, U, ".--", markersize=12, lw=1)
    plt.xlabel(r"Энергия кванта $h\nu$, мэВ")
    plt.ylabel(r"Сигнал фотоответа $\Delta \Sigma/N_0$")
    plt.savefig(f"output//A-{style}.png")
    
    plt.clf()
    plt.subplots_adjust(bottom=0.15)
    plt.plot(hv, U**.5, ".--", markersize=12, lw=1)
    plt.xlabel(r"Энергия кванта $h\nu$, мэВ")
    plt.ylabel(r"Сигнал фотоответа $(\Delta U/N_0)^{1/2}$")
    plt.savefig(f"output//B-{style}.png")
```
