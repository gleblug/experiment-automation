```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 51
t = np.linspace(0, 50e-9, l)
```

```python
omega1 = 2e8
phi1 = 1
a1 = 0.6
omega2 = 2e8
phi2 = 2
a2 = 0.4
```

```python
V1 = a1*np.sin(omega1*t + phi1)
V2 = a2*np.sin(omega2*t + phi2)
```

```python
plt.plot(t, V1)
plt.plot(t, V2)
plt.show()
```
